```python
#3. Now we will try to check the normality of the numerical variables visually
# Use seaborn library to construct distribution plots for the numerical variables

import seaborn as sns
fig, axis = plt.subplots(2,4, figsize=(12,7))    
sns.histplot(df['customer_lifetime_value'].values, kde=True, ax = axis[0,0]).set_xlabel('customer_lifetime_value')
sns.histplot(df['income'].values, kde=True, ax = axis[0,1]).set_xlabel('income')
sns.histplot(df['monthly_premium_auto'].values, kde=True, ax = axis[0,2]).set_xlabel('monthly_premium_auto')
sns.histplot(df['months_since_last_claim'].values, kde=True, ax = axis[0,3]).set_xlabel('months_since_last_claim')
sns.histplot(df['months_since_policy_inception'].values, kde=True, ax = axis[1,0]).set_xlabel('months_since_policy_inception')
sns.histplot(df['number_of_open_complaints'].values, kde=True, ax = axis[1,1]).set_xlabel('number_of_open_complaints')
sns.histplot(df['number_of_policies'].values, kde=True, ax = axis[1,2]).set_xlabel('number_of_policies')
sns.histplot(df['total_claim_amount'].values, kde=True, ax = axis[1,3]).set_xlabel('total_claim_amount')

plt.tight_layout()
```
